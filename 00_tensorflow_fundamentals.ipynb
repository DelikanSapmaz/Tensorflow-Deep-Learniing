### Creating random tensors
Raandom tesnors are tensores of same abitrary size random numbers(Yeni rasgele tensorler rasgele sayılar içeren bazı keyfi boyutlardaki zamanlardır)

```python
# create two random (but the same) tensors
random_1=tf.random.Generator.from_seed(7)
random_1=random_1.normal(shape=(3,2))
#Bu kod, TensorFlow kütüphanesinin random modülünü kullanarak, normal dağılıma göre rastgele sayılar üretir.

#İlk satırda, rastgele sayıların üretiminde kullanılacak olan random_1 adlı bir nesne oluşturulur ve 42 sayısı ile başlatılır. Bu nesne, TensorFlow'un random modülünden türetilmiştir.

#İkinci satırda, normal() metodu kullanılarak, random_1 nesnesinden 3 satır ve 2 sütundan oluşan bir dizi üretilir. Bu dizi, normal dağılıma göre rastgele sayılar içerir. Bu sayılar, her seferinde farklı olacaktır, ancak 42 başlangıç değeri sabit kalacaktır.

#Sonuç olarak, random_1 adlı değişkende 3 satır ve 2 sütundan oluşan bir dizi saklanacaktır, ve bu dizi normal dağılıma göre rastgele sayılar içerecektir.
random_2=tf.random.Generator.from_seed(7)
random_2=random_2.normal(shape=(3,2))

# Are they equal?
random_1,random_2,random_1==random_2
```

**The dot product**

Matrix çarpım olarak adlandırılır

You can perform matrix multiplication using:

* tf.matmul()

* tf.tensordot(),

* @

```python
X,Y
```

```python
# Perfrom the dot product on X  and Y (requires X or Y to be be transposed)
tf.tensordot(tf.transpose(X),Y,axes=1)
# Burada, axes=1 parametresi, A ve B tensörlerinin sütunları ile satırları arasında bir eşleştirme yapılacağını belirtir. C tensörü, A ve B tensörlerinin nokta çarpımını temsil eder.

# Bu şekilde, tf.matmul() fonksiyonu matrisler arasındaki doğrudan bir çarpma işlemi gerçekleştirirken, tf.tensordot() fonksiyonu nokta çarpımı yapmak için iki tensörün şekillerini eşleştirir.
```

```python
# Perform matrix mutliplication between X and Y (transposed)
tf.matmul(X,tf.transpose(Y))
```

```python
# Perform matrix mutliplication between X and Y (reshaped)
tf.matmul(X,tf.reshape(Y,shape=(2,3)))
```

transpose işlemi ise, tensörün boyutlarını yer değiştirir. Matrislerde transpoze, matrisin satır ve sütunlarının yer değiştirmesi anlamına gelir. transpose işlemi, tensörün elemanlarının dizilimini değiştirmez, ancak tensörün şeklini değiştirir.

Bu nedenle, reshape ve transpose işlemleri farklı sonuçlar verirler ve farklı amaçlar için kullanılırlar. reshape işlemi tensörün şeklini değiştirirken, transpose işlemi tensörün boyutlarını değiştirir.

```python
# Check values of Y, reshape Y and tranposed Y
print("Normal Y:")
print(Y, "\n") # "\n" for newline

print("Y reshaped to (2, 3):")
print(tf.reshape(Y, (2, 3)), "\n")

print("Y transposed:")
print(tf.transpose(Y))
```

GPU (Graphics Processing Unit - Grafik İşlem Birimi), bilgisayarların grafik işlemlerini hızlandırmak için kullanılan bir donanım bileşenidir. GPU'lar, özellikle oyunlar ve video işleme gibi yoğun grafik gerektiren uygulamalarda yüksek performans sağlarlar. Ayrıca, yapay zeka ve derin öğrenme gibi yoğun hesaplama gerektiren işlemlerde de kullanılabilirler.

TPU (Tensor Processing Unit - Tensor İşlem Birimi), Google tarafından geliştirilmiş bir özel işlem birimidir. TPU'lar, yapay zeka ve derin öğrenme gibi büyük ölçekli veri işleme işlerinde özel olarak tasarlanmıştır. TPU'lar, özellikle Google'ın yapay zeka ve makine öğrenmesi platformu olan TensorFlow ile kullanıldığında yüksek performanslı hesaplama yapabilmektedirler. TPU'lar, büyük ölçekli veri işleme işleri için özellikle yüksek performans sağlar ve CPU ve GPU'lardan daha yüksek bir performans sunarlar.
